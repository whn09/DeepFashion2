```python
import cv2
import os
import json
from json2xml import json2xml

classes_cnt = {}

def cocojson2vocxml(img, filename):
    data = json.load(open(filename, 'r'))
#     print('data:', data)
    objects = []
    for i in range(1, 11):  # TODO assume max 10 items
        item_name = 'item'+str(i)
        if item_name in data:
            annotation = data[item_name]
            name = annotation['category_name'].lower()
            xmin = int(annotation['bounding_box'][0])
            if xmin <= 0:
                xmin = 1
            ymin = int(annotation['bounding_box'][1])
            if ymin <= 0:
                ymin = 1
            xmax = int(annotation['bounding_box'][2])
            ymax = int(annotation['bounding_box'][3])
            obj = {
                'name': name, 
                'pose': 'Unspecified', 
                'truncated': 0, 
                'difficult': 0, 
                'bndbox': {'xmin': xmin, 'ymin': ymin, 'xmax': xmax, 'ymax': ymax}}
            objects.append(obj)
            if name not in classes_cnt:
                classes_cnt[name] = 0
            classes_cnt[name] += 1
    new_data = {
        'folder': '', 
        'filename': filename.split('/')[-1], 
        'path': '', 
        'source': {}, 
        'size': {'width': img.shape[1], 'height': img.shape[0], 'depth': 3}, 
        'object': 'TODO'
    }
    # print(new_data)
    xml = json2xml.Json2xml(new_data, wrapper='annotation', pretty=True, attr_type=False).to_xml()
    objects_xml = ''
    for obj in objects:
        object_xml = json2xml.Json2xml(obj, wrapper='object', pretty=True, attr_type=False).to_xml()
        objects_xml += object_xml.replace('<?xml version="1.0" ?>', '')+'\n'
    xml = xml.replace('<object>TODO</object>', objects_xml)
    with open(filename.replace('.json', '.xml'), 'w') as fout:
        fout.write(xml)

def coco2voc(dirname):
    filenames = os.listdir(os.path.join(dirname, 'image'))
    for i, filename in enumerate(filenames):
        if i % 1000 == 0:
            print(dirname, ':', i, '/', len(filenames))
        if filename.endswith('.jpg'):
            img = cv2.imread(os.path.join(dirname, 'image', filename))
            if img is None:
                print('img error:', os.path.join(dirname, 'image', filename))
            else:
                cocojson2vocxml(img, os.path.join(dirname, 'annos', filename.replace('jpg', 'json')))
    print(classes_cnt)
    print(classes_cnt.keys())
```

```python
coco2voc('data/tiny_train')
```

```python
train_dir = 'data/train'
validation_dir = 'data/validation'
coco2voc(train_dir)
coco2voc(validation_dir)
```
